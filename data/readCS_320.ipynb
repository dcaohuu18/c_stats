```python
def create_connection(db):
    conn = None
    try:
        conn = sqlite3.connect(db)
    except Error as e:
        print(e)

    return conn
```

```python
def read_course(conn,term_code):
    cur = conn.cursor()
    cur.execute("""SELECT *
                   FROM '"""+term_code+"""'
                   WHERE term_code = '"""+term_code+"""';""")

    rows= cur.fetchall()
    column_names = [col[0] for col in cur.description]
    df = pd.DataFrame(rows, columns= column_names)
    print(df)
```
